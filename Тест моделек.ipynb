Записываем списки числовых и категориальных признаков еще раз(почему то он не подтягивает его (?)), затем кодируем их через гетдаммис и указываем dtype int(без этого проставит true or false и все равно не сможем с ними работать), затем создаем таблицу х и склеиваем наши числовые и закодированые признаки (axis=1 означает, что он склеивает таблицы иначе он бы просто их расширил вниз и удалил первую часть), затем для у выносим колонку с ушел или нет она для нас основная, и просто печатаем таблицу чтобы узнать ее размерность и с помощью head визуализируем ее

```python
cat_cols = [	
    'Sex',	
    'IsSeniorCitizen',	'HasPartner',	'HasChild',	'HasPhoneService',	
    'HasMultiplePhoneNumbers',	'HasInternetService',	'HasOnlineSecurityService',	'HasOnlineBackup',	
    'HasDeviceProtection',	'HasTechSupportAccess',	'HasOnlineTV',	'HasMovieSubscription',	
    'HasContractPhone', 'IsBillingPaperless',
    'PaymentMethod' 
]

num_cols = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent'
]

df=df.dropna()



dummy_features = pd.get_dummies(df[cat_cols], dtype=int)

X = pd.concat([df[num_cols], dummy_features], axis=1)

y = df['Churn']

print(f"Таблица готова, размер: {X.shape}")
X.head(10)
```

импортируем из sklearn для тестов и для стандартизации числовых признаков, затем разбераем на обучение и тест (0,2 значит 20% оставим на тест) и рандомное значение ставим чтобы в будущем был такой же результат, даем переменной функцию стандартизатора, на трейне через фит трансформ приводим данные к нормальному ввиду (считаем среднее и отклонение,,затем проходится по каждому отнимает от значения среднее и делит на отклонение тем самым получает поинты типо +1,2 или -0,8) теперь моделки могут легко с этим работать (!! Важно мы на трейни делали фит трансформ поскольку фит считал данные,а трансформ уже подгонял с их помощью,на тесте мы используем только трансформ,то есть мы через старые данные даем поинты новым иначе мы тупо пересчитали все по новому),затем выводим трейн и тест и видим, что все преобразовалось нормально

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

print(X_train['TotalSpent'].head(5))
print(X_test['TotalSpent'].head(5))
```

Создаем логрегресию (!! ошибся в рок ауке запомнить,что точность по обычным предиктам (0 и 1), а рок аук по вероятностям), импортируем из склерна модель и метрики, даем переменной модель затем учим модельку на наших данных, потом идет переменная с жесткими предиктами и переменная с предиктами вероятности и по ним уже строим переменные с метриками (!!Обязательно в accuracy используем жесткие предикты,а в рок ауке вероятности, первое поломается,а второе даст результат хуже (до было 0,73)), затем выводим их и окргуляем

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,accuracy_score

model = LogisticRegression(C=1, random_state=19)  

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

y_pred_proba = model.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_pred_proba)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")
```

Строим модель ближайших соседей, импортируем и даем переменной (указал смотреть на 5 получилось 0,77, затем перебором 115 показала наилучшей результат (?? тут нужно подбирать гиперпараметр, возможно организовать жадный поиск,попробовать с логрегресией еще (как повлияет C на метрики?))), точно так жем учим модельку и даем предикты жесткие и вероятности, !!дотянуть до 0.85 не получается нужно будет вернуться 

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=115)

knn.fit(X_train, y_train)

knn_pred_probs = knn.predict_proba(X_test)[:, 1] 
knn_pred = knn.predict(X_test)             

knn_roc_auc = roc_auc_score(y_test, knn_pred_probs)
knn_acc = accuracy_score(y_test, knn_pred)

print(f"Accuracy: {knn_acc:.4f}")
print(f"ROC-AUC: {knn_roc_auc:.4f}")
```

Теперь можем построить для наших 3 моделек матрицу ошибок (TP,TN,FP,FN), импортируем затем создаем список наших моделек в котором лежат кортежи пар, имя модели и сама обученая модель, затем создаем холст где у нас 1 линия и 3 колонки так же и выдаем размер, затем циклом идем по списку (enumerate позволяет пронумеровать ), name и clf это пара (название модели и сама модель), заетм даем переменной импортируемое значение с .from_estimator (прямо из модели брать) он берет на каждой итерации модельку и данные и она сравнивает предикты на этом строя матрицу, затем задаем чтобы питон рисовал в номере по итерации,а не рандомно и выдаем цвет окраса (может быть любой,красный подошел лучше всего), затем просто берем из пробы колонку с предиктами и сравниваем со значениями теста и выдаем метрику roc-auc, затем просто подписываем это все, по номеру подписать имя модели потом на новой строчке значение roc-auc округленное до 4 после запятой и выводим это все

```python
from sklearn.metrics import ConfusionMatrixDisplay

models_list = [
    ('logregresion', model),
    ('knn', knn),
    ('svm', svm_model)
]

fig, axes = plt.subplots(1, 3, figsize=(20, 5))


for i, (name, clf) in enumerate(models_list):
    
    cm_display = ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, ax=axes[i], cmap='Reds')
    
    score = roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1])
    axes[i].set_title(f"{name}\nROC-AUC: {score:.4f}")

plt.show()
```

Сделаем подбор с лучшей моделью catboost для этого импортирую optuna и logging (второе нужно только для того что бы выключить сообщения о поборе на каждой попытке), затем создаем функцию к которой будем обращаться каждую итерацию в ней прописываем параметры которые будем перебирать (trial.suggest_вариант подбора может быть целым с плавующей точкой или даже строчкой для подбора допустим ядра в методе опорных векторов), так же для рассмотра маленьких значений нужно поставить log=True тк для того же лр это очень важно и может сильно повлиять на результат, так же можно указать параметры которые будут закреплены и оптюна перебирать их уже не будет (например зерно, и 'allow_writing_files': False тк на каждой итерации кетбуст в папку с проектом кладет кучу файлов отчета и это для нас бесполезно так что лучше это вырубить, так же неплохо указать по какой метрике мы идем,хоть мы вычисляем ее дальше,но это должно улучшать внутринее вычисления), потом мы в модельки можем с помощью ** распаковывать словарь с параметрами сразу в модель,обучать ее и получать результат,он возвращается в оптюну и она его записывает и продолжает попытки, !! важно что параметр в журнале study_cb = optuna.create_study(direction='maximize') мы ставим максимизацию нашей метрики, если бы была ошибка то мы могли бы поставить минимизацию, затем просто запускаем это все параметр  n_trials отвечает за количество подборов, а show_progress_bar выводит удобную шкалу сколько уже завершенно и какая попытка и метрика были наилучшими, в конце выводим результат и параметры, (можно сделать вывод,что оптюна себя очень хорошо показала результат довльно заметно улучшился, в принципе зная гиперпараметры для подбора можно подобрать их для любой модели !!важно не напутать вид гиперпараметра (int или float или str), а так же указать минимизацию или максиммизацию метрики)

```python
import optuna
import logging



optuna.logging.set_verbosity(optuna.logging.WARNING)

def objective(trial):
    params = {
        
        'n_estimators': trial.suggest_int('n_estimators', 300, 1000),
        'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.1, log=True),
        'depth': trial.suggest_int('depth', 3, 10),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10),
        'random_strength': trial.suggest_float('random_strength', 0, 10),
        
        
        'random_state': 19,
        'verbose': 0,
        'allow_writing_files': False,
        'eval_metric': 'AUC'
    }
    
    model = CatBoostClassifier(**params)
    model.fit(X_train, y_train)
    
    probs = model.predict_proba(X_test)[:, 1]
    return roc_auc_score(y_test, probs)

study_cb = optuna.create_study(direction='maximize')


study_cb.optimize(objective, n_trials=500, show_progress_bar=True)



print(f"лучший результат: {study_cb.best_value:.4f}")
print("Лучшие параметры:", study_cb.best_params)
```
